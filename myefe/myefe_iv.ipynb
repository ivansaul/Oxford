```python
import pandas as pd
```

```python
df=pd.read_excel('Irregular_Verbs_Myefe.xls')
# drop by Name
df = df.drop(['Meaning'], axis=1)
df
```

```python
def split_word(s):
    s=s.replace('*','')
    i=s.find('[')
    return [s[:i].strip().lower(), s[i:].strip()]

def clear_notes(s):
    if isinstance(s,str):
        s=s.replace(u'\xa0',u' ').replace(u'&#8211;',u'-')
    else:
        s=None
    return s
```

```python
df['Infinitive']=df['Infinitive'].apply(split_word)
df['Past Simple']=df['Past Simple'].apply(split_word)
df['Past Participle']=df['Past Participle'].apply(split_word)
df['Notes']=df['Notes'].apply(clear_notes)
df['word']=df['Infinitive'].apply(lambda x: x[0])
```

```python
df.sample(n=10)
```

```python
df.to_json('myefe_iv.json', orient='records', indent=4)
```
