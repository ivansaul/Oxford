# **Merge data oxford-apk-myefe**

```python
import pandas as pd
```

### **Read oxford data**

```python
df_ox=pd.read_csv('oxford.csv')
df_ox['word']=df_ox['word'].str.lower()
# drop by uk_mp3 - us_mp3
df_ox = df_ox.drop(['level','uk_mp3','us_mp3'], axis=1)
#add two new columns
df_ox['pron'] = pd.Series(dtype='object')
df_ox['sense'] = pd.Series(dtype='object')
df_ox['infinitive'] = pd.Series(dtype='object')
df_ox['past simple'] = pd.Series(dtype='object')
df_ox['past participle'] = pd.Series(dtype='object')
df_ox['notes'] = pd.Series(dtype='object')
df_ox
```

### **Read apk data**

```python
df_enes=pd.read_json('apk/en-es.json')
df_enes['word']=df_enes['word'].str.lower()
df_enes.set_index('word',inplace=True)
df_enes
```

### **Read myefe irregular verbs**

```python
df_iv=pd.read_json('myefe/myefe_iv.json')
df_iv['word']=df_iv['word'].str.lower()
df_iv.set_index('word',inplace=True)
df_iv
```

### **Merge dataframes**

```python
for ind in df_ox.index:
    try:
        word=df_ox['word'][ind]
        sr_enes=df_enes.loc[word] #df series
    except:
        pass
        #print(ind, '--', df_ox['word'][ind])
    else:
        pron=sr_enes['pron']
        sense=sr_enes['sense']
        #set new values
        df_ox.at[ind, 'pron'] = pron
        df_ox.at[ind, 'sense'] = sense
        #print(ind, '--',word)
        try:
            sr_iv=df_iv.loc[word] #df series
        except:
            pass
        else:
            inf=sr_iv['Infinitive']
            pasim=sr_iv['Past Simple']
            papar=sr_iv['Past Participle']
            notes=sr_iv['Notes']
            #set new values
            df_ox.at[ind, 'infinitive'] = inf
            df_ox.at[ind, 'past simple'] = pasim
            df_ox.at[ind, 'past participle'] = papar
            df_ox.at[ind, 'notes'] = notes 
```

### **Cleaning Data**

```python
df_ox.dropna(axis='index',how='any',subset=['pron','sense'], inplace=True)
df_ox
```

### **Save new data as json file**

```python
df_ox.to_json('merge.json', orient='records', indent=4)
```
